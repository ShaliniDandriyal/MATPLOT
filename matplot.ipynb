```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",align='right',color='b',label="ml students")
plt.hist(py_students_age,bins,rwidth=0.8,histtype="bar",align='right',color='y',label="py students")
plt.title("ML DATA")
plt.xlabel("Students age")
plt.ylabel("No. of Students")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,histtype="bar",align='right',color=['b','y'],label=["ml students","py_students"])
plt.title("ML DATA")
plt.xlabel("Students age")
plt.ylabel("No. of Students")
plt.legend()
plt.show()
```
